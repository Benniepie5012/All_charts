```python
flights = sn.load_dataset('flights')
flights.head()
```

```python
#Matrix plot
plt.figure(figsize=(8,6))
sn.set_context('paper', font_scale=1.4)
```

```python
flights = flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
flights
```

```python
plt.figure(figsize=(12,8))
sn.heatmap(flights, cmap='Blues', linecolor='white', linewidth=1);
```

```python
#cluster map
sn.clustermap(flights, cmap='Blues', standard_scale=1)
```
