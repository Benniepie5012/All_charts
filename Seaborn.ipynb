```python
iris = sn.load_dataset('iris')
iris.head()
```

```python
iris.species.value_counts()
```

```python
sn.set_style('darkgrid')
sn.scatterplot(data=iris, x='sepal_length', y='petal_length', hue='species');
```

```python
iris.corr()
```

```python
mat_iris = iris.corr(method='pearson')
```

```python
sn.heatmap(mat_iris, annot=True)
```

```python
sn.scatterplot(data=iris, x='sepal_length', y='petal_length', hue='species', size='species');
```

```python
sn.pairplot(iris);
```

```python
sn.pairplot(iris, hue='species', palette="Set1")
```

```python
#lm plot
sn.lmplot(x='petal_length', y = 'petal_width', hue= 'species', markers=["o", "*", "^"], data = iris)
```

```python
#linear relationship with different variable
sn.lmplot(x='petal_length', y='petal_width', hue='species', data=iris, markers=["o", "^", "*"]);
```

```python
sn.boxplot(x='species', y= 'sepal_width', data=iris);
```
